# Exploratory Data Analysis

This notebook explores the dataset used for fine-tuning.


```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load data
with open('../data/processed/train.json', 'r') as f:
    train_data = json.load(f)

df = pd.DataFrame(train_data)
print(f"Dataset size: {len(df)}")
print(f"\nColumns: {df.columns.tolist()}")
df.head()
```

```python
# Analyze text lengths
df['instruction_length'] = df['instruction'].str.len()
df['output_length'] = df['output'].str.len()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['instruction_length'].hist(ax=axes[0], bins=50)
axes[0].set_title('Instruction Length Distribution')
df['output_length'].hist(ax=axes[1], bins=50)
axes[1].set_title('Output Length Distribution')
plt.tight_layout()
plt.show()
```
